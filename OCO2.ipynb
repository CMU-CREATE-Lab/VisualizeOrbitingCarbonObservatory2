V6R:  higher quality, uses forward data

We recommend using nadir data with warn levels equal to or less than 8, and glint data with warn levels less than or equal to 11. Currently, due to unexpected bad spectral samples, we also recommend using a chi^2 filter of 5 in the weak band and 4 in the strong band.

```python
import futures, h5py, os, re, sys, urllib2
```

```python
def mirrorFile(src, dest):
    if os.path.exists(dest):
        print '%s already exists, skipping' % dest
        return
    try:
        os.makedirs(os.path.dirname(dest))
    except:
        pass
    print 'Reading %s' % src
    sys.stdout.flush()
    body = urllib2.urlopen(src).read()
    open(dest + '.tmp', 'w').write(body)
    os.rename(dest + '.tmp', dest)
    print 'Mirrored %.1f MB from %s to %s' % (len(body) / 1e6, src, dest)

src = "ftp://oco2.gesdisc.eosdis.nasa.gov/data/s4pa/OCO2_DATA/OCO2_L2_Standard.7r/2014/257/oco2_L2StdGL_01074a_140914_B7000r_150715094513.h5"
dest = "OCO2_L2_Standard.7r/2014/257/oco2_L2StdGL_01074a_140914_B7000r_150715094513.h5"

mirrorFile(src, dest)
```

```python
#src = 'ftp://oco2.gesdisc.eosdis.nasa.gov/data/s4pa/OCO2_DATA/OCO2_L2_Standard.7r/2014/257'
src = 'http://oco2.gesdisc.eosdis.nasa.gov/opendap/OCO2_L2_Standard.7r/2014/257'
dest = 'OCO2_L2_Standard.7r/2014/257'
def mirrorDir(src, dest):
    try:
        body = urllib2.urlopen(src + '/').read()
    except urllib2.HTTPError, e:
        if e.code == 404:
            print '%s not found' % src
            return
        raise e
    files = sorted(list(set([match.group(1) for match in re.finditer(r'"(\w+\.h5)"', body)])))
    print 'Found %d files in %s' % (len(files), src)
    sys.stdout.flush()
    with futures.ThreadPoolExecutor(max_workers=6) as e:
        for file in files:
            e.submit(mirrorFile, src + '/' + file, dest + '/' + file)

mirrorDir(src, dest)
```

```python
def mirrorAll():
    for year in range(2014, 2016):
        for day in range(1, 367):
            dest = 'OCO2_L2_Standard.7r/%d/%d' % (year, day)
            src = 'http://oco2.gesdisc.eosdis.nasa.gov/opendap/' + dest
            mirrorDir(src, dest)
            
#mirrorAll
```
